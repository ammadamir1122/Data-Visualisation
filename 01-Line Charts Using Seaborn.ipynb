```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
spotify_Data = pd.read_csv("spotify.csv", index_col="Date", parse_dates=True)
```

```python
spotify_Data.head()
```

```python
spotify_Data.tail()
```

```python
sns.lineplot(data = spotify_Data)
```

```python
plt.figure(figsize=(14,6))

plt.title("Daily stream of popular songs in 2017-2018")

sns.lineplot(data = spotify_Data)
```

```python
list(spotify_Data.columns)
```

```python
plt.figure(figsize=(14, 6))

plt.title("Daily stream of poppular songs in 2017-2018")

sns.lineplot(data = spotify_Data['Shape of You'], label = "Shape of You")

sns.lineplot(data = spotify_Data['Despacito'], label = "Despacito")

plt.xlabel("Date")
plt.ylabel("Stream Frequency")
```
