```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
insurance_data = pd.read_csv("insurance.csv")
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x = insurance_data["bmi"], y = insurance_data["charges"])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
#checking the relationship of BMI and insurance charges to the smokers and non-smoker
#we can see that with the increasing bmi smokers pay more charges, then the non-smokers

sns.scatterplot(x = insurance_data['bmi'], y = insurance_data['charges'], hue = insurance_data['smoker'])
```

```python
#plotted two regression lines for smokers and non smokers
#We can clearly see theat the slope of smokers regression line is much more steeper relative to the line for non-smokers
sns.lmplot(x = 'bmi', y = 'charges', hue = 'smoker', data = insurance_data)
```

```python
sns.swarmplot(x = insurance_data['smoker'], y = insurance_data['charges'])
```
